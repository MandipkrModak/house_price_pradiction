```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/housing price.csv")
data
```

```python
data.info()
```

```python
data.dropna(inplace = True)
data.info()
```

```python

from sklearn.model_selection import train_test_split

x = data.drop(["median_house_value"], axis = 1)

y = data["median_house_value"]

```

```python
x_train , x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)

```

```python
y_train
```

```python
train_data = x_train.join(y_train)
train_data
```

```python
train_data.hist(figsize= (13,8),color="Green")
```

```python
plt.figure(figsize = (13,8))
sns.heatmap(train_data.corr(numeric_only=True),annot= True , cmap = "YlGnBu")

```

```python
train_data["total_rooms"] = np.log(train_data["total_rooms"] +1)
train_data["total_bedrooms"] = np.log(train_data["total_bedrooms"] +1)
train_data["population"] = np.log(train_data["population"] +1)
train_data["households"] = np.log(train_data["households"] +1)
```

```python
train_data.hist(figsize = (15,8),color = "Green")
```

```python
train_data.ocean_proximity.value_counts()
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(["ocean_proximity"], axis = 1)

plt.figure(figsize = (13,8))
sns.heatmap(train_data.corr(numeric_only=True),annot= True , cmap = "YlGnBu")
```

```python
plt.figure(figsize = (13,8))
sns.scatterplot(x ="latitude", y = "longitude", data = train_data, hue = "median_house_value", palette = "coolwarm")
```

```python
train_data["household_rooms"] = train_data["total_rooms"]/train_data["households"]
train_data["bedrooms_ratio"] = train_data["total_bedrooms"]/train_data["total_rooms"]

```

```python
plt.figure(figsize = (13,8))
sns.heatmap(train_data.corr(numeric_only=True),annot= True , cmap = "YlGnBu")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train, y_train = train_data.drop(["median_house_value"], axis = 1), train_data["median_house_value"]
x_train_s = scaler.fit_transform(x_train)

rag = LinearRegression()

rag.fit(x_train_s, y_train)
```

```python
test_data = x_test.join(y_test)


test_data["total_rooms"] = np.log(test_data["total_rooms"] +1)
test_data["total_bedrooms"] = np.log(test_data["total_bedrooms"] +1)
test_data["population"] = np.log(test_data["population"] +1)
test_data["households"] = np.log(test_data["households"] +1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(["ocean_proximity"], axis = 1)

test_data["household_rooms"] = test_data["total_rooms"]/test_data["households"]
test_data["bedrooms_ratio"] = test_data["total_bedrooms"]/test_data["total_rooms"]
```

```python
test_data.head()
```

```python
x_test, y_test = test_data.drop(["median_house_value"], axis = 1), test_data["median_house_value"]
```

```python
x_test_s = scaler.transform(x_test)
```

```python
rag.score(x_test_s, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

forest  = RandomForestRegressor()
forest.fit(x_train_s, y_train)

```

```python
forest.score(x_test_s, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Create the RandomForestRegressor model
forest = RandomForestRegressor(random_state=42)

# Define the parameter grid
param_grid = {
    "n_estimators": [10, 20, 30],
    "min_samples_split": [2, 4],
    "max_depth": [None, 6, 8],
}

# Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=forest,
    param_grid=param_grid,
    cv=5,  # 5-fold cross-validation
    scoring="neg_mean_squared_error",
    return_train_score=True,
    n_jobs=-1  # Use all processors for parallel computation
)

# Fit the GridSearchCV object
grid_search.fit(x_train_s, y_train)

# Output the best estimator
print("Best Estimator:", grid_search.best_estimator_)
```

```python
grid_search.best_estimator_
```

```python
best_rf_model = grid_search.best_estimator_
y_pred = best_rf_model.predict(x_test_s)
accuracy = best_rf_model.score(x_test_s, y_test)
```
